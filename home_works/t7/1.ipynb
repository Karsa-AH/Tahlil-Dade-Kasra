```python
import math

def sort_points_by_distance(points):
    """Sorts points based on their distance from the origin (0,0)."""
    return sorted(points, key=lambda p: math.sqrt(p[0]**2 + p[1]**2))


points = [(3, 4), (1, 1), (0, 0), (-2, -3), (5, 12)]
sorted_points = sort_points_by_distance(points)

print(sorted_points)
```
