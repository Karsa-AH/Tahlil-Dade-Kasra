```python
import requests
from bs4 import BeautifulSoup
import sys
import sqlite3
conn = sqlite3.connect('iranjib.db')
cursor = conn.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS news (
        id INTEGER PRIMARY KEY,
        title TEXT,
        url TEXT,
        date TEXT,
        summary TEXT
    )
''')
conn.commit()

id = 0
for i in range(200):
    base= "https://www.iranjib.ir/showarchive/?/0/"
    x=base
    x=x.replace("?",str(i))
    rsite = requests.get(x)
    bsite = BeautifulSoup(rsite.content , 'html.parser')
    ftitels=bsite.find_all("tr")
    ftitels = ftitels[1:]
    titles = []
    for tr in ftitels:
        tds = tr.find_all('td')
        if len(tds) == 3 and tds[0].find('a'):  # Only news rows have 3 tds and an <a> tag in first <td>
            titles.append(tr)
    for news in titles:
        c = news.find("a")
        title=c.get_text() 
        id += 1
        if " ارز " in title or "طلا" in title:
            date=news.find_all("td")[2]
            e=date.get_text()
            if "۱۴۰۴" in e:

                
        
                url=c.get("href")
                lsite=requests.get(url)
                lsite=BeautifulSoup(lsite.content,"html.parser")
                summery=lsite.find_all("div",class_="matn")
                summery=summery[0].get_text()
                body = ""  # placeholder since body is not extracted

                insert_statement = '''INSERT INTO news (id, title, summary, url, date) 
                                    VALUES ( ?, ?, ?, ?, ?)'''
                cursor.execute(insert_statement, (id, title, summery, url, e))
                conn.commit()


                print(f"id = {id} ---\ntitles = {title} ---\n urls = {url} ---\n date = {date} ---\n summery = {summery}\n")

            else:
                conn.commit()
                sys.exit("all of the new were found")
```
